```python
!pip install yfinance==0.2.4
import yfinance as yf
import matplotlib.pyplot as plt

def make_graph(ticker_symbol, title):

    stock_data = yf.download(ticker_symbol, start='2021-01-01', end='2022-01-01')

    plt.figure(figsize=(10, 6))
    plt.plot(stock_data['Close'], label='Close Price')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Close Price (USD)')
    plt.legend()
    plt.grid(True)
    plt.show()

ticker_symbol = 'TSLA'
graph_title = 'Tesla Stock Data'

make_graph(ticker_symbol, graph_title)
```
